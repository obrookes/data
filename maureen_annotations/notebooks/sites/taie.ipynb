```python
import pandas as pd
from glob import glob
```

```python
old_platform = pd.read_csv('candSID_PanAfID_translations_old_platform_2021_11_27.csv')
old_platform['subdir'] = old_platform['new.file.name'].str.split('/').str[-2].str.lower().str.rstrip('+')
old_platform['video'] = old_platform['new.file.name'].str.split('/').str[-1].str.split('.').str[0].str.lower()
old_platform['subdir_video'] = old_platform['subdir'] + '/' + old_platform['video']
```
