```python
import pandas as pd
```

```python
adiko_df = pd.read_csv("cotedivoire.csv")
adiko_df["yes_reaction"] = adiko_df["no_reaction"].apply(lambda x: False if x else True)
```

```python
adiko_df.avoidance.value_counts(), adiko_df.attraction.value_counts(), adiko_df.inspection.value_counts(), adiko_df.no_reaction.value_counts()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 5))
adiko_df[["avoidance", "attraction", "inspection", "no_reaction"]].sum().plot(
    kind="bar"
)
ax.set_ylabel("Number of Videos")
ax.set_xlabel("Behaviour")
ax.set_title("Tai Camera Reaction Behaviour: Multilabel")
# rotate x labels
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")
plt.savefig("tai_camera_reaction_behaviour.png", bbox_inches="tight")
```

```python
adiko_df[["yes_reaction", "no_reaction"]].sum().plot(kind="bar")
ax.set_ylabel("Number of Videos")
ax.set_xlabel("Behaviour")
ax.set_title("Tai Camera Reaction Behaviour: Binary")
# rotate x labels
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")
```

```python
guinea_ne_df = pd.read_csv("guinea_ne.csv")
guinea_nw_df = pd.read_csv("guinea_nw.csv")
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
guinea_ne_df[["avoidance", "attraction", "inspection", "no_reaction"]].sum().plot(
    kind="bar"
)
# Log y axis
ax.set_ylabel("Number of Videos")
ax.set_xlabel("Behaviour")
ax.set_title("Guinea NE Camera Reaction Behaviour: Multilabel")
# rotate x labels
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")
# plt.savefig("tai_camera_reaction_behaviour.png", bbox_inches="tight")
```

```python
fig, ax = plt.subplots(figsize=(12, 5))
width = 0.25
guinea_ne_df[["avoidance", "attraction", "inspection", "no_reaction"]].sum().plot(
    kind="bar", position=0, width=width, ax=ax
)
guinea_nw_df[["avoidance", "attraction", "inspection", "no_reaction"]].sum().plot(
    kind="bar", position=1, width=width, ax=ax, color="darkorange"
)
# log y axis
ax.set_yscale("log")
ax.set_ylabel("Number of Videos")
ax.set_xlabel("Behaviour")
ax.set_title("Guinea Camera Reaction Behaviour: Multilabel")
ax.legend(["North East", "North West"])
# rotate x labels
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")
plt.savefig("guinea_camera_reaction_behaviour.png", bbox_inches="tight")
```
